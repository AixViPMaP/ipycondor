# Customizable job table

```python
from ipycondor import Condor
from classad import ExprTree
import time
ui=Condor()
ui.job_table( str(ExprTree('JobStartDate') > int(time.time())-6*3600), 
             ['JobStartDate','QDate','Owner','JobStatus','RemoteHost','DiskUsage']
)
```

# Get ClassAd objects of jobs

```python
from ipycondor.Condor import deep_parse
for job in deep_parse(ui.jobs()):
    print(job['ClusterId'], job['JobStatus'])
```
